Tuples, Dictionaries, and Sets in Python

```python
# Tuples
# we can't append anything in tuple, once created
a = "mango","apple",7,8.19
b = ("mango","apple",7,8.19,"banana")
print(type(a))
print(type(b))
print(a)
c = ("Ironman")
print(type(c))
# to convert c to tuple
d = ("Ironman",)
print(type(d))
```

```python
# slicing and iterations in tuples
x = ("vivo","redmi","oppo","apple","nokia")
print(x)
print(x[1:3])
# to print 1st 3 values
print(x[:3])
print(x[2:]) # to print values from 3rd till last
# to gap 1 in between
print(x[::2])
print(x[1::2]) # to gap 1 in between starting from redmi
print(x[::-1]) # to print in reverse order
print(x[2::-1]) # to print in reverse starting from 2nd val
```

```python
# Conversion of Tuples and Tuple Functions
x = ("vivo","redmi","oppo","apple","nokia")
print(x)
print(type(x))
x = list(x)
print(x)
print(type(x))
x.append("micromax")
print(x)
x = tuple(x)
print(x)
print(type(x))
```

```python
# to check count of element
print(x.count("redmi"))
print(x.index("redmi")) # to check index of elem
```
