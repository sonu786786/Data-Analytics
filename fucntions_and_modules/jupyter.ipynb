# Python Functions and Modules

```python
def hello():
  print("hello world")
hello()
```

```python
# parameters and arguments
def add(x,y): # x,y are parameters
  print(x+y)
add(29,87) # 29, 37 are arguments
add(49,49)
```

```python
def rectArea(length,breadth):
  print("area of rectangle is: ",length*breadth,"unit sq")
rectArea(4,6)
```

```python
# arbitrary argument
def hello(*name):
  print("hello! my name is",name[1])
hello("sonu","aman","ankit")
```

```python
# Return statement and recursion
def hello():
  return "hello world"

print(hello())
hello()
```

```python
def add(x,y):
  return x+y
print("addition of two no.s: ",add(44,32))
```

```python
# finding factorial
def factorial(x):
  if x==1:
    return 1
  return x*factorial(x-1)
print(factorial(4))
```

```python
# Lambda fuction
a = lambda b: b*4
print(a(4))
```

```python
x = lambda a,b,c: (a+b)*c
print(x(4,3,8))
```
